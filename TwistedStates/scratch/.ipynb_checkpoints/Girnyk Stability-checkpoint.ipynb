# Stability Comparison
This compares the results of using equation (3.6) of Medvedev & Tang (2015) wiht Figure 1 of Girnyk, Hasler, & Maistrenko (2012).

I am finding an inconsistency.



Figure 1 from Girnyk, Hasler, Maistrenko (2012).  Left is attractive, right is repulsive.

```julia
using Plots
using TwistedStates
using Printf
```

## Attractive Case

```julia
k_vals = 1:n÷2 - 1;
q_vals = 0:n-1;
n = 20;
α = 0;
kk = [];
qq = [];
for k in k_vals, q in q_vals
    λ = get_q_twisted_spectrum(n, q, k, α)
    if maximum(λ[2:end])<0
        push!(kk, k);
        push!(qq, q);
    end
end

scatter(kk, qq, label="")
xlims!(0, n÷2)
ylims!(0, n)
title!("Attractive")
xlabel!("k")
ylabel!("q")
```

The above figure differs from the one from the Girnyk paper.  For example, the Girnky paper predicts that the $q =3$ and $k=2$ state should be stable here.  Howeve,r if we compute the spectrum for this state, we find that it has positive eigenvalues.

```julia
get_q_twisted_spectrum(n, 3, 2, 0)
```

## Repulsive Case

```julia
k_vals = 1:n÷2 - 1;
q_vals = 0:n-1;
n = 20;
α = 1;
kk = [];
qq = [];
for k in k_vals, q in q_vals
    λ = get_q_twisted_spectrum(n, q, k, α)
    if maximum(λ[2:end])<0
        push!(kk, k);
        push!(qq, q);
    end
end

scatter(kk, qq, label="")
xlims!(0, n÷2)
ylims!(0, n)
title!("Repulsive")
xlabel!("k")
ylabel!("q")
```

Again, we see some disagreement.  For intsance, $q=5$ and $k=2$ should be stable, but:

```julia
get_q_twisted_spectrum(n, 5, 2, 1)
```

This is probably stable, as the single positive eigenvalue is on hte scale of floating point.  This may correspond to a second neutral mode.  Supose instead we look at $q=6$ and $k=6$:

```julia
get_q_twisted_spectrum(n, 6, 6, 1)
```

This predicts an unstable state with positive eigenvalues that are unambiguous.
